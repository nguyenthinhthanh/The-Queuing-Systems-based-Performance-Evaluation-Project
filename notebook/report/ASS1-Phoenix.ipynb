### 8.2 Mapping the output simulation measurement and mathematical formula calculation
The table **below** presents the correspondence between the **measured performance metrics** obtained from the MLFQ simulation and the **analytical parameters** derived from the M/M/m queueing theory model. The mapping table bridges the gap between the **MLFQ simulation results** and the **theoretical M/M/m queueing model**, ensuring that every measured parameter has a clear analytical counterpart for performance analysis and validation.

| Simulation Output | Formula | Description |
|--------------------------|:----------------:|-------------|
| `throughput` | `λ_eff = served_cpu / sim_time` | Effective throughput (jobs completed per unit time) |
| `cpu_service_mean` | `1 / μ` | Average service time per completed job |
| `cpu_util` | `ρ = λ / (m · μ)` | CPU utilization (fraction of busy cores) |
| `avg_turnaround` | `W` | Average turnaround / response time |
| `avg_wait_per_job` | `Wq`| Average waiting time per job |
| `generated` | `Arrivals` | Total number of jobs generated |

**Table 2. Simulation & mathematical formula mapping table**

In particular:

- **`throughput (λ_eff)`** represents the **effective job completion rate**, calculated as the number of completed jobs divided by the total simulation time. It corresponds to the analytical system throughput `λ_eff`.
- **`cpu_service_mean (1/μ)`** indicates the **average service time per completed job**, representing the reciprocal of the service rate `μ`.
- **`cpu_util (ρ)`** measures the **CPU utilization**, equivalent to the theoretical traffic intensity `ρ = λ / (m · μ)`.
- **`avg_turnaround (W)`** reflects the **average turnaround time**, also known as the **response time**, defined as the total time from job arrival to completion.
- **`avg_wait_per_job (Wq)`** approximates the **average waiting time** experienced by a job in the ready queues before being served.
- **`generated`** and **`dropped`** indicate the total number of jobs generated and the number of jobs dropped (if system capacity is limited), respectively.
